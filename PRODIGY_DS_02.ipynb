```python
df = sns.load_dataset("titanic")
df
```

```python
df["age"]=df["age"].fillna(28)
```

```python
df.isna().sum()
```

```python
df["age"]=df["age"].astype(int)
df
```

```python
df["fare"]=df["fare"].round(2)
df
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('who').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
plt.hist(df['age'])
plt.xlabel('Age')
plt.ylabel('Number of passengers')
plt.title('Distribution of passenger ages')
plt.show()
```

```python
# Group data by class and calculate the mean fare
df_grouped = df.groupby('Class')['fare'].mean()

# Prepare data for the bar chart
class_names = df_grouped.index.to_list()
average_fares = df_grouped.values.tolist()

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(class_names, average_fares)

# Add labels and title
plt.xlabel('Class')
plt.ylabel('Average Fare')
plt.title('Average Fare')
```

```python
df['age_group'] = pd.cut(df['age'], bins=[0, 18, 35, 60, 100], labels=['Children', 'Young Adults', 'Adults', 'Elderly'])

age_counts = df['age_group'].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(age_counts, labels=age_counts.index, autopct='%1.1f%%')
plt.title('Age Distribution of Titanic Passengers')
plt.show()
```
